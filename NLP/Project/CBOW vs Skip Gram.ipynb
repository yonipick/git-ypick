```python
import pandas as pd
import numpy as np
%matplotlib inline
from __future__ import print_function
import collections
import math
import os
import random
import tensorflow as tf
#import zipfile
from matplotlib import pylab
import matplotlib.pyplot as plt
from six.moves import range
from six.moves.urllib.request import urlretrieve
from sklearn.manifold import TSNE
import gensim
```

```python
# Importing the dataset
dataset = pd.read_csv('GameOfThrons.txt',header=None, delimiter = '\n', quoting = 3)
dataset.rename(columns={"0": "Lines_Numbers"})
dataset.head()
```

```python
# Cleaning the texts
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
#from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, dataset.size):
    line = re.sub('[^a-zA-Z]', ' ', dataset[0][i])
    line = line.lower()
    line = line.split()
    #ps = PorterStemmer()
    line = [word for word in line if not word in set(stopwords.words('english'))]
    line = ' '.join(line)
    corpus.append(line)
```

```python
corpus.remove('')
len(corpus)
```

```python
print(dataset[0][4])
print(corpus[4])
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()

```

```python
print (len(cv.vocabulary_))
#print(cv.get_feature_names())
```

```python
data = tf.keras.preprocessing.sequence.skipgrams(sequence=dataset,vocabulary_size=len(cv.vocabulary_),negative_samples=5,
                                          window_size=4, seed =5)
```
